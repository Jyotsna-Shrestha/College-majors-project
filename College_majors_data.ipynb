# **Cleaning data:**

```python
df.isna()
```

```python
df.tail()
```

```python
df_clean = df.dropna()
```

```python
df_clean.tail()
```

# **Finding the maximum and minimum paying majors:**

```python
max_salary_id = df_clean['Starting Median Salary'].idxmax()
print(f"The major with the highest starting salary is {df_clean['Undergraduate Major'].loc[max_salary_id]} and the starting salary is ${round(df_clean['Starting Median Salary'].max())}.")
```

```python
max_midsalary_id = df_clean['Mid-Career Median Salary'].idxmax()
print(f"The major with the highest mid-career salary is {df_clean['Undergraduate Major'].loc[max_midsalary_id]} and the salary is ${round(df_clean['Mid-Career Median Salary'].max())}.")
```

```python
min_salary_id = df_clean['Starting Median Salary'].idxmin()
print(f"The major with the lowest starting salary is {df_clean['Undergraduate Major'].loc[min_salary_id]} and the starting salary is ${round(df_clean['Starting Median Salary'].min())}.")
```

```python
min_midsalary_id = df_clean['Mid-Career Median Salary'].idxmin()
print(f"The major with the lowest mid-career salary is {df_clean['Undergraduate Major'].loc[min_midsalary_id]} and the salary is ${round(df_clean['Mid-Career Median Salary'].min())}.")
```

# **Finding low risk major:**

```python
spread = df_clean['Mid-Career 90th Percentile Salary'] - df_clean['Mid-Career 10th Percentile Salary']
df_clean.insert(1, 'Spread', spread)
low_risk = df_clean.sort_values('Spread')
low_risk_major = low_risk['Undergraduate Major'].loc[low_risk['Spread'].idxmin()]
print(f"Low risk major is {low_risk_major}.")
```

# **Finding majors with highest potential:**

```python
high_potential = df_clean.sort_values('Mid-Career 90th Percentile Salary',ascending=False)
high_potential.head()
```
